```python
from keras.models import load_model
from collections import deque
import numpy as np
import pickle
import cv2
```

```python
lb = pickle.loads(open(r"/Users/priyankafrancis/Desktop/Dissertation/output/video2.pickle",'rb').read())
model = load_model(r"/Users/priyankafrancis/Desktop/Dissertation/Video_Classification_model/videoclassification")
outputVideo = (r"/Users/priyankafrancis/Desktop/Dissertation/Output/outputvideo/demo3_output.avi")
mean = np.array([123.68, 116.779, 103,939][:1] , dtype="float32")
Queue = deque(maxlen=128)
```

```python
import cv2
import numpy as np

capture_video = cv2.VideoCapture("/Users/priyankafrancis/Desktop/Dissertation/Video_Classification_model/demo4.mp4")
writer = None
(Width, Height) = (None, None)
Queue = []  # Initialize the Queue

while True:
    (taken, frame) = capture_video.read()
    if not taken:
        break
    if Width is None or Height is None:
        (Width, Height) = frame.shape[:2]

    output = frame.copy()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame = cv2.resize(frame, (244, 224)).astype("float32")  # Adjust height to 224
    frame = frame - mean  # Assuming 'mean' is defined somewhere

    # Ensure the correct input shape for the model
    frame = np.expand_dims(frame, axis=0)

    preds = model.predict(frame)[0]
    Queue.append(preds)
    results = np.array(Queue).mean(axis=0)
    i = np.argmax(results)
    label = lb.classes_[i]
    text = "They are playing: {} Video, Technique: ResNet50".format(label)
    cv2.putText(output, text, (45, 60), cv2.FONT_HERSHEY_SIMPLEX, 1.25, (19, 237, 77), 5)

    if writer is None:
        fourcc = cv2.VideoWriter_fourcc(*"MJPG")
        writer = cv2.VideoWriter("outputVideo.avi", fourcc, 30, (Width, Height), True)
    writer.write(output)
    cv2.imshow("In progress", output)
    key = cv2.waitKey(1) & 0xFF

    if key == ord("q"):
        break

print("finalizing.......")
writer.release()
capture_video.release()
```
